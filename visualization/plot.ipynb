```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
df = pd.read_csv('../data/simulation_data.csv')

# Plot data for each researcher individually
researcher_ids = df['ID'].unique()

# Plot Knowledge for all researchers on the same plot
plt.figure(figsize=(10, 8))

researcher_ids = df['ID'].unique()
for researcher_id in researcher_ids:
    researcher_data = df[df['ID'] == researcher_id]
    plt.plot(researcher_data['Time'], researcher_data['InnovationScore'], label=f'Researcher {researcher_id}')

# Adding title and labels
plt.title('Knowledge Over Time for All Researchers')
plt.xlabel('Time')
plt.ylabel('Innovation Score')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
